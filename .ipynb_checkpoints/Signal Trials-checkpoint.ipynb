```python
file1 = sio.loadmat(r"PPG_ACC_dataset/S1/squat1_acc.mat")
type(file1["ACC"])
t=file1["ACC"][:,0]
x=file1["ACC"][:,1]
y=file1["ACC"][:,2]
z=file1["ACC"][:,3]
file2 = sio.loadmat(r"PPG_ACC_dataset/S1/squat1_ppg.mat")
ppg=(file2["PPG"][:,1])


def butter_highpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = signal.butter(order, normal_cutoff, btype='high', analog=False)
    return b, a

def butter_highpass_filter(data, cutoff, fs, order=5):
    b, a = butter_highpass(cutoff, fs, order=order)
    y = signal.filtfilt(b, a, data)
    return y

def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = signal.butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = signal.filtfilt(b, a, data)
    return y


ppg_filtered1=butter_highpass_filter(ppg,3,1000)
ppg_filtered2=butter_lowpass_filter(ppg_filtered1,10,1000)


plt.figure(figsize=[30,20])


subplot(3,1,1)
plt.plot(t,x)
plt.plot(t,y)
plt.plot(t,z)
title('ACC_Squat_1', fontsize=20)
plt.xlabel('time (s)', fontsize=18)
plt.ylabel('amplitude mV', fontsize=16)
plt.grid()

subplot(3,1,2)
plt.plot(t,ppg_filtered2)
plt.grid()
plt.show()
```
